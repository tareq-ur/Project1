```python
def normalize(col):
 mean = col.mean()
 min = col.min()
 max = col.max()
 std = col.std()
 return (col - min )/ (min - max)
add['id'].agg([normalize])
```

```python
def standarize(col):
 mean = col.mean()
 min = col.min()
 max = col.max()
 std = col.std()
 return (col - mean )/ std
add['id'].agg([standarize])
```

```python
import numpy as np
import pandas as pd
import os
```

```python
os.getcwd()
root_dir = os.path.dirname(os.getcwd())
root_dir
data_dir = os.path.join(root_dir, 'Data_set')
data_dir
data_set = os.path.join(data_dir, 'sales Data.csv')
data_set

print("root_dir: ", root_dir)
print("data_dir: ", data_dir)    
print("data_set: ", data_set)
```

```python
sales_data =pd.read_csv(data_set)
sales_data.head(100)
```

```python
# Remove irrelevant columns 
if 'Unnamed: 0' in sales_data.columns:
    sales_data.drop(columns=['Unnamed: 0'], inplace=True)
sales_data.head(10)
```

```python
sales_data['weekly_sales'].describe()
```

```python
sales_data['weekly_sales'].mean()
print("Mean: ", sales_data['weekly_sales'].mean())
sales_data['weekly_sales'].median()
print("Median: ", sales_data['weekly_sales'].median())
sales_data['weekly_sales'].std()
print("Standard Deviation: ", sales_data['weekly_sales'].std())
sales_data['weekly_sales'].max()
print("Max: ", sales_data['weekly_sales'].max())
sales_data['weekly_sales'].min()
print("Min: ", sales_data['weekly_sales'].min())
sales_data['weekly_sales'].quantile(0.25)
print("25th Percentile: ", sales_data['weekly_sales'].quantile(0.25))
sales_data['weekly_sales'].quantile(0.75)
print("75th Percentile: ", sales_data['weekly_sales'].quantile(0.75))
```

```python
def normalize (col):
    mean = col.mean()
    std = col.std()
    return (col - mean) / std
sales_data['weekly_sales'].agg([normalize])
sales_data.head(10)
```
